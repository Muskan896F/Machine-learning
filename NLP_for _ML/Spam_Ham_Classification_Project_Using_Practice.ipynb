### **Data Cleaning And Preprocessing**

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = [ ]
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-z]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Output Features
y = pd.get_dummies(messages['label'])
```

```python
y = y.iloc[:,0].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train),len(y_train)
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=300,linewidth=100000,formatter=dict(float=lambda x : "%.3g" % x))
```

```python
X_train
```

```python
X_test
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# **TF-IDF**

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X_train = tfidf.fit_transform(X_train).toarray()
X_test = tfidf.fit_transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
spam_tfidf_model
```

```python
y_pred = spam_tfidf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
